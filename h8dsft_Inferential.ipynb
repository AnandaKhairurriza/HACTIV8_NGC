### Import Library

```python
import pandas as pd
import numpy as np
import matplotlib as mat
from scipy import stats
```

### Read data csv

```python
df = pd.read_csv("data/covid_19_indonesia_time_series_all.csv")
df
```

```python
# New case per hari di Jakarta
case_Jakarta = df[df["Location"] == "DKI Jakarta"][["Date", "New Cases"]].groupby("Date").sum()
# New case per hari di Jawa Barat
case_JB = df[df["Location"] == "Jawa Barat"][["Date", "New Cases"]].groupby("Date").sum()

print(f'Rata - rata new case Jakarta : {round(case_Jakarta["New Cases"].mean(),2)}')
print(f'Rata - rata new case Jawa Barat : {round(case_JB["New Cases"].mean(),2)}')
```

Rata - rata new case Jakarta : 1520.42

Rata - rata new case Jawa Barat : 1266.16

H0: μ_Jakarta = μ_JawaBarat

H1: μ_Jakarta != μ_JawaBarat

Apakah rata-rata new case Jakarta tidak ada perbedaan signifikan pada rata-rata new case Jawa Barat atau terdapat perbedaan yang signifikan

### Pembuktian dengan single sample two sided

```python
t_stat, p_val = stats.ttest_ind(case_Jakarta,case_JB)
print('P-value:',p_val[0])
print('t-statistics:',t_stat[0])
```

Hasil P-value adalah 0.02 yaitu dibawah critical value (0.05), maka hipotesis null yang menyatakan tidak ada perbedaan signifikan dapat ditolak. Sehingga rata-rata new case Jakarta dan Jawa Barat berbeda cukup signifikan.
