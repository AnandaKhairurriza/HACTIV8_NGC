### Soal 2

```python
a = [17, 22, 19]
b = [10, 20, 11]
c = [5, 12, 9]

mat_a = np.array(a)
mat_b = np.array(b)
mat_c = np.array(c)

# A + B
print(f"A + B = {mat_a + mat_b}")

# B - C
print(f"B - C = {mat_b - mat_c}")

# A . C
print(f"A . C = {mat_a @ mat_c}")

# A x B
print(f"A X B = {mat_a * mat_b}")

# norm A
normMan = np.linalg.norm(mat_a, 1)
normEu = np.linalg.norm(mat_a, 2)
print(f"norm A (Manhattan) = {normMan}") #Manhattan
print(f"norm A (Euclar) = {normEu}") #Euclar

# sudut A dan B
cos_theta = (mat_a @ mat_b) / (np.linalg.norm(mat_a) * np.linalg.norm(mat_b))
print(f"Sudut antara A dan B = {cos_theta}")
```

COMING SOON (Tunggu sampai week 4!)

```python
#TEMPORARY
import matplotlib.pyplot as plt

u = [2,5]
v = [3,1]
mat_u = np.array(u)
mat_v = np.array(v)
origin_point = np.array([[0,0], [0,0]])

plt.quiver(*origin_point, mat_u, mat_v, color=['r','g'], scale=21)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```
