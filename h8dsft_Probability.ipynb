Buatlah sebuah kode untuk mensimulasikan melempar koin sebanyak 50.000 kali dengan assign value random berupa 0 dan 1 untuk kepala atau ekor. Setiap kali koin dilempar, peluang untuk mendapatkan kepala atau ekor adalah 50%.

```python
%matplotlib inline
import random as rd
from matplotlib import pyplot as plt
from scipy import special as sps
```

```python
koin = [0,0] #Isi 2 value karena variasi hanya 2 (kepala dan ekor)

percobaan = 50000 #dilempar 50000 kali
cobake = 0

while cobake < percobaan:
    x = rd.randint(0,1) #Generate integer random (0 sampai 1)
    koin[x] += 1 #Jika x = 0 masuk ke jumlah kepala, x = 1 masuk ke ekor
    cobake += 1
    
print(koin) #Print jumlah muncul kepala dan ekor
```

```python
plt.figure(figsize=(6,6)) #Ukuran chart
plt.pie(koin, labels=['kepala', 'ekor'])
plt.legend() #Menampilkan informasi (biru = kepala, kuning = ekor)
plt.show()
```
